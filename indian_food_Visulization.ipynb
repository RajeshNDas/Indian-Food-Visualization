# **Data Visualization**

---



```python
plt.rcParams['figure.figsize'] = (10,8)
sns.countplot(ind_fd['course'])
plt.title("<Type of Course Numbers>")
plt.show()
```

**Checking all course type counts, we can see there majorities of main course in the Dataframe**

```python
plt.rcParams['figure.figsize'] = (10,8)
sns.countplot(ind_fd['flavor_profile'],palette=("rocket"))
plt.title("<Type of Flavours counts>")
plt.show()
```

**Visualization of all the flavour counts, We can see clearly that spicy food are maximum.**

```python
plt.rcParams['figure.figsize'] = (10,8)
sns.countplot(ind_fd['diet'],palette=("Set2"))
plt.title("<Type of Diets>")
plt.show()
```

***Visualization of the diet, There are more vegetarian food than non-veg ***

```python
plt.rcParams['figure.figsize'] = (15,8)
sns.countplot(ind_fd['prep_time'],palette=("hls"))
plt.title("<Preparation Times>")
plt.show()
```

**Visualization of the preparation time of these foods, Max foods are short prep time.**

```python
plt.rcParams['figure.figsize'] = (15,8)
sns.countplot(ind_fd['cook_time'],palette=("mako"))
plt.title("<Cooking Time>")
plt.show()
```

**Visualization of Cook time of the foods, Majority foods have 30.0 cook time.**

## **Indepth Visualizations**

---



```python
plt.rcParams['figure.figsize'] = (30,20) #size of the plot
sns.countplot(ind_fd['state'],hue = 'flavor_profile', data = ind_fd,palette=("Paired"))
plt.title("<STATE WISE FLAVORS>")
plt.show()
```

**Visualization of Flavours with respect to states. We can see punjab has most spicy foods, West bangal has most no of sweet dishes and only Gujrat has sour dishes.**

```python
plt.rcParams['figure.figsize'] = (18,15)
sns.scatterplot(data=ind_fd, x="prep_time", y="state", style="prep_time")
```

**Scatterplot Visualization prep time with respect of State**

```python
plt.rcParams['figure.figsize'] = (18,15)
sns.scatterplot(data=ind_fd, x="cook_time", y="state", style="cook_time")
```

**Scatterplot Visualization of cook time with respect of State**

```python
plt.rcParams['figure.figsize'] = (30,20)
sns.countplot(ind_fd['state'],hue = 'diet', data = ind_fd, palette=('Set2'))
plt.title("<STATE WISE DIET>")
plt.show()
```

**Visualization of Diet with respect of State. Gujrat has maximum no of vegetarian dishes and assam has maximum no of non vegetarian dishes**

```python
plt.rcParams['figure.figsize'] = (30,20)
sns.countplot(ind_fd['state'],hue = 'course', data =ind_fd)
plt.title("<STATE WISE COURSE>")
plt.show()
```

**Visualization of course with respect to State, Punjab has maximum number of main course compared to other states, West bengal has maximum number of Deserts, Only punjab has starters, and Gujrat has Max number of snacks**

```python
pd.crosstab(ind_fd["state"],ind_fd["flavor_profile"]).plot(kind="bar",figsize=(20,15))
plt.title("<STATE WISE MOST LIKED FLAVOR>")
plt.show()
```

**Showing Statewise most liked flavour. Where we can see punjab likes spicy dishes than other flavours, West bengal likes sweet dishes most, only ounjab and gujrat has bitter dishes, and gujrat has sour dishes.**

```python
pd.crosstab(ind_fd["state"],ind_fd["course"]).plot(kind="bar",figsize=(20,15))
plt.title("<STATE WISE MOST LIKED COURSE>")
plt.show()
```

**Showing statewise most liked course, Punjab liked Main course more then any state, West bengal's most liked course is desert, Gujrat likes Snack most.**

```python
pd.crosstab(ind_fd["state"],ind_fd["diet"]).plot(kind="bar",figsize=(20,15))
plt.title("<STATE WISE DIET>")
plt.show()
```

**Showing Most liked diet statewise, Gujrat has maximum number of Veg dishes, and Assam has maximum number of non veg dishes.**

```python
pd.crosstab(ind_fd["region"],ind_fd["flavor_profile"]).plot(kind="bar",figsize=(20,15))
plt.title("<REGION WISE MOST LIKED FLAVOR>")
plt.show()
```

**Showing Region wise most liked flavour, In west region Spicy sour and sweet flavour liked most.**

```python
pd.crosstab(ind_fd["region"],ind_fd["course"]).plot(kind="bar",figsize=(20,15))
plt.title("<REGION WISE MOST LIKED COURSE>")
plt.show()
```

**Region wise liked course. North has spicy as most liked course. West has most snack as most liked course and also desert.**

```python
plt.rcParams['figure.figsize'] = (18,15)
plt.title("<COOK TIME BASED ON REGION>")
sns.scatterplot(data=ind_fd, x="cook_time", y="region", style="cook_time")
```

**Showing Cook time based on Region.**

```python
plt.rcParams['figure.figsize'] = (18,15)
plt.title("<PREP TIME BASED ON REGION>")
sns.scatterplot(data=ind_fd, x="prep_time", y="region", style="cook_time")
```
